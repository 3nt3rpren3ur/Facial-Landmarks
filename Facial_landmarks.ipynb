# **Facial landmarks detection for images**

```
import imageio
import matplotlib.pyplot as plt
from mlxtend.image import extract_face_landmarks
import numpy as np

img = imageio.imread('me1.jpg')
landmarks = extract_face_landmarks(img)
print(landmarks.shape)
```

```
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot(1, 3, 1)
ax.imshow(img)
ax = fig.add_subplot(1, 3, 2)
ax.scatter(landmarks[:, 0], -landmarks[:, 1], alpha=0.8)
ax = fig.add_subplot(1, 3, 3)
img2 = img.copy()

for p in landmarks:
    img2[p[1]-2:p[1]+2, p[0]-2:p[0]+2, :] = (255, 255, 255)


ax.imshow(img2)
plt.show()
```

```
import matplotlib.pyplot as plt
%matplotlib inline

left_eye = np.array([36, 37, 38, 39, 40, 41])
right_eye = np.array([42, 43, 44, 45, 46, 47])
fig = plt.figure(figsize=(10,10))
plt.plot(landmarks[:,0], -landmarks[:,1], 'o')
for i in range(landmarks.shape[0]):
    plt.text(landmarks[i,0]+1, -landmarks[i,1], str(i), size=14)


left_eye_center = np.mean(landmarks[left_eye], axis=0)
right_eye_center = np.mean(landmarks[right_eye], axis=0)

plt.plot([left_eye_center[0]], [-left_eye_center[1]], marker='p', color='blue', markersize=10)

plt.plot([right_eye_center[0]], [-right_eye_center[1]], marker='p', color='blue', markersize=10)

plt.show()
```
